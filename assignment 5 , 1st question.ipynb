```python
##Residual Analysis
###Test for Normality of Residuals (Q-Q Plot)
```

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-10000)) 
```

```python
#####Residual Plot for Homoscedasticity
```

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

```python
#####Model Deletion Diagnostics
###Detecting Influencers/Outliers
####Cook’s Distance
```

```python
final_ml_rd_a= smf.ols('Profit~rdspend+Administration',data = startup).fit()
(final_ml_rd_a.rsquared,final_ml_rd_a.aic)
```

```python
final_ml_rd_m= smf.ols('Profit~rdspend+marketingspend',data = startup).fit()
(final_ml_rd_m.rsquared,final_ml_rd_m.aic)
```

```python
#Comparing above R-Square and AIC values, model 'final_ml_m' has high R- square and low AIC value hence include variable 'marketingspend' so that multi collinearity problem would be resolved.
```

```python
model_influence_V = final_ml_m.get_influence()
(c_V, _) = model_influence_V.cooks_distance
```

```python
fig= plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(startup)),np.round(c_V,3));
plt.xlabel('Row index')
plt.ylabel('Cooks Distance');
```

```python
#Since the value is <1 , we can stop the diagnostic process and finalize the model
```

```python
final_ml_m= smf.ols('Profit~rdspend+marketingspend',data = startup).fit()
(final_ml_m.rsquared,final_ml_m.aic)
```
