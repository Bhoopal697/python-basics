```python
stats.norm.cdf(-0.02,loc = mba_df.gmat.mean(),scale= mba_df.gmat.std())# to find the normal distribution 
```

```python
mba_df_ci= stats.norm.interval(0.95,loc=mba_df.gmat.mean(),scale=mba_df.gmat.std()) #to find the confidence interval 
print( 'Gmat score at 95% confidence interval is:', np.round(mba_df_ci, 2))
```
